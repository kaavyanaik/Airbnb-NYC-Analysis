```python
# File: 04_eda_airbnb.ipynb
# Purpose: Perform visual EDA using the cleaned dataset

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('cleaned_airbnb.csv')

# Plot 1: Room type distribution
plt.figure(figsize=(6, 4))
df['room_type'].value_counts().plot(kind='barh', color='skyblue')
plt.title('Room Type Distribution')
plt.tight_layout()
plt.savefig('room_type_distribution.png')
plt.show()
```

```python
# Plot 2: Average price by borough
plt.figure(figsize=(8, 5))
sns.barplot(data=df, x='neighbourhood_group', y='price', estimator='mean')
plt.title('Average Price by Borough')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('rice_by_borough.png')
plt.show()
```

```python
# Plot 3: Price vs Availability
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='availability_365', y='price', hue='room_type', alpha=0.6)
plt.title('Price vs Availability by Room Type')
plt.tight_layout()
plt.savefig('price_vs_availability.png')
plt.show()
```
